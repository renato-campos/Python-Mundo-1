Exercício 1: Listar a **pasta_0**, que está dentro da pasta **files**, caso existam arquivos **.txt** dentro da **pasta_0**, pintar o conteúdo dos arquivos.

```python
from pathlib import Path
```

```python
caminho = Path()/'files'
caminho.mkdir()
```

```python
caminho = caminho / 'pasta_0'
caminho.mkdir()
```

```python
for i in range(10):
    nome = caminho / f'arquivo{i}.txt'
    nome.touch()
```

```python
lista = list(caminho.glob('*.txt'))
```

```python
# abrindo os arquivos via 'with' e escrevendo
for arq in lista:
    with open(arq, 'w') as f:
        f.write(f'{arq}')
```

```python
# abrindo e lendo os arquivos agora via 'glob' com seu iterador de resposta
for f in Path('files/pasta_0').glob('*.txt'):
    print(f.read_text())
```

Exercício 2: listar a pasta **pasta_0**, que está dentro da pasta **files**, caso existam arquivos **.txt** dentro da **pasta_0**, crie uma nova pasta chamada **desafio_2** e copie os arquivos para essa pasta.

```python
caminho = Path() / 'files'/'desafio_2'
caminho.mkdir()
for f in Path('files/pasta_0').glob('*.txt'):
    #shutil.move(f, caminho)
    shutil.copy(f, caminho)
```
